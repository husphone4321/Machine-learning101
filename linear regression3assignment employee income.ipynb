Here is a simple example of linear regression data for employee income based on years of experience:

| Employee | Years of Experience | Income (in thousands $) |
|----------|---------------------|-------------------------|
| A        | 1                   | 50                      |
| B        | 2                   | 60                      |
| C        | 3                   | 70                      |
| D        | 4                   | 80                      |
| E        | 5                   | 90                      |

In this table:
- "Years of Experience" is the independent variable (x-axis).
- "Income" is the dependent variable (y-axis).

The goal of the linear regression would be to find the relationship between years of experience and income, which in this case appears to be a straight line with an increase of 10 thousand dollars per year of experience. 

This is a simplified real-world scenario where typically more factors might influence income and the relationship may not always be perfectly linear.

```python
import pandas as pd
import numpy as np
# Initialize the data
data = {
    'Employee': ['A', 'B', 'C', 'D', 'E'],
    'Years of Experience': [1, 2, 3, 4, 5],
    'Income (in thousands $)': [50, 60, 70, 80, 90]
}
# Define the standard deviation of the noise
noise_std_dev = 5

# Generate random noise
noise = np.random.normal(0, noise_std_dev, len(data['Income (in thousands $)']))

# Add noise to the income data
data['Income (in thousands $)'] += noise

# Create a DataFrame
df = pd.DataFrame(data)
# Print the DataFrame
print(df)


```

Visualize the data using matplotlib:

```python
import matplotlib.pyplot as plt

# Plotting the data
plt.figure(figsize=(8, 6))
plt.scatter(df['Years of Experience'], df['Income (in thousands $)'], color='blue', label='Data Points')
plt.title('Employee Income vs Years of Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Income (in thousands $)')
plt.grid(True)
plt.legend()
plt.show()
```

fit a linear regression model to predict employee income based on various features such as years of experience, education level, job title, and location.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Prepare the data for the model
X = df[['Years of Experience']]  # Feature (independent variable)
y = df['Income (in thousands $)']  # Target (dependent variable)

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)
```

visualize the fitted line and the residuals

```python
# Plotting the data and the fitted line
plt.figure(figsize=(8, 6))
plt.scatter(df['Years of Experience'], df['Income (in thousands $)'], color='blue', label='Noisy Data Points')
plt.plot(df['Years of Experience'], model.predict(X), color='red', label='Fitted Line')
plt.title('Employee Income vs Years of Experience with Noise and Fitted Line')
plt.xlabel('Years of Experience')
plt.ylabel('Income (in thousands $)')
plt.grid(True)
plt.legend()
plt.show()
```

print model coefficients
print model intercept

```python
#print model coefficients and intercept
print("Model coefficients: ", model.coef_)
print("Model intercept: ", model.intercept_)    
```

To calculate the Sum of Squares Regression (SSR) for your linear regression model, you need to follow these steps:

1. **Calculate the predicted values from the model.**
2. **Calculate the mean of the actual values.**
3. **Compute the SSR using the formula:**

\[ \text{SSR} = \sum (\hat{y}_i - \bar{y})^2 \]

Where:
- \(\hat{y}_i\) is the predicted value for the \(i\)-th observation.
- \(\bar{y}\) is the mean of the actual values.

Here's how you can do it in your Jupyter Notebook:


```python
# Predict the values
y_pred = model.predict(X)

# Calculate the mean of the actual values
y_mean = np.mean(y)

# Calculate the SSR
ssr = np.sum((y_pred - y_mean) ** 2)

print(f"Sum of Squares Regression (SSR): {ssr}")
```

Certainly! Here is the additional code to calculate the \( R^2 \) value of your linear regression model:


You can add this code snippet to your existing notebook to calculate and print the \( R^2 \) value.

```python

# Calculate the R^2 value
r_squared = model.score(X, y)

print(f"R^2 (Coefficient of Determination): {r_squared}")
```
